```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def read_intarray(filename):
    data = []
    with open(filename, 'r') as fh:
        for l in fh:
            l = l.strip()
            data.append([int(c) for c in l])
    return np.array(data)

test = read_intarray("test.txt")
test
```

```python
def isvisible(d, i, j):
    h, w = d.shape
    if i == 0 or i == h-1 or j == 0 or j == w-1:
        return 1
    x = d[i,j]
    for line in d[0:i,j], d[i+1:,j], d[i,0:j], d[i,j+1:]:
        if np.all(np.less(line, x)):
            return 1
    return 0

def scenic(d, i, j):
    h, w = d.shape
    if i == 0 or i == h-1 or j == 0 or j == w-1:
        return 0
    x = d[i,j]
    out = []
    for line in reversed(d[0:i,j]), d[i+i,j], reversed(d[i,0:j]), d[i,j+1]:
        out.append(0)
        for e in line:
            out[-1] += 1
            if e >= x:
                break
    print(out)    

def allvisible(d):
    h, w = d.shape
    out = []
    for i in range(h):
        out.append([isvisible(d, i, j) for j in range(w)])
    return np.array(out)

testvis = allvisible(test)
plt.matshow(test)
plt.matshow(testvis)
print(np.sum(testvis))
scenic(test, 1, 2)
```
